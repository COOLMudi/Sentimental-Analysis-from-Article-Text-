```python
import re

def listToString(s): 
    # initialize an empty string
    str1 = "" 
    # traverse in the string  
    for ele in s: 
        str1 += ele  
    # return string  
    return str1 

def clean_text(x):
  x=listToString(x)
  x = x.lower()
  x = x.encode('ascii', 'ignore').decode()
  x = re.sub("@\S+", " ", x)
  x = re.sub("https*\S+", " ", x)
  x = re.sub("#\S+", " ", x)
  x = re.sub("\'\w+", '', x)
  x = re.sub('\s{2,}', " ", x)
  return x

d['TEXT']=d['TEXT'].apply(clean_text)
d['TEXT'].loc[0]
```

```python
import pandas as pd
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import re
nltk.download('stopwords')
```

```python
lemma = WordNetLemmatizer()
stop_words = stopwords.words('english')
```

### `Text Preprocessing : Cleaning`

---



```python
def text_prep(x):
     corp = str(x).lower() 
     corp = re.sub('[^a-zA-Z]+',' ', corp).strip() 
     tokens = word_tokenize(corp)
     words = [t for t in tokens if t not in stop_words]
     lemmatize = [lemma.lemmatize(w) for w in words]
    
     return lemmatize
```

```python
nltk.download('punkt')
preprocess_tag = [text_prep(i) for i in d['TEXT']]
d["preprocess_txt"] = preprocess_tag
d['total_len'] = d['preprocess_txt'].map(lambda x: len(x))
d.head()
```

```python
import textstat
from textstat.textstat import textstatistics
import spacy

def syllables_count(word):
    return textstatistics().syllable_count(word)

def difficult_words(text):

    words = text
    # difficult words are those with syllables >= 2
    # easy_word_set is provide by Textstat as
    # a list of common words
    diff_words_set = set()
     
    for word in words:
        syllable_count = syllables_count(word)
        if word not in stop_words and syllable_count >= 2:
            #print(word)
            diff_words_set.add(word)
 
    return len(diff_words_set)
d[' Complex_Words Count']= d['preprocess_txt'].apply(difficult_words)
d['Percentage of Complex_Words']=(( d['preprocess_txt'].apply(difficult_words))/x1)*100
```

# `Personal_Pronoun`

```python
import re
def Personal_Pronoun(x):
  pronounRegex = re.compile(r'\b(I|we|my|ours|(?-i:us))\b',re.I)
  pronouns = pronounRegex.findall(x)
  return pronouns

d['Personal Pronoun']=d['TEXT'].apply(Personal_Pronoun)
```

# `avg_word_length`

```python
def avg_word_length(x):
  words = x.split()
  average = sum(len(word) for word in words) / len(words)
  return average

d['Avg Word Length']=d['TEXT'].apply(avg_word_length)
```

# `Syllables Word`

```python
import re

VOWEL_RUNS = re.compile("[aeiouy]+", flags=re.I)
EXCEPTIONS = re.compile(
    # fixes trailing e issues:
    # smite, scared
    "[^aeiou]e[sd]?$|"
    # fixes adverbs:
    # nicely
    + "[^e]ely$",
    flags=re.I
)
ADDITIONAL = re.compile(
    # fixes incorrect subtractions from exceptions:
    # smile, scarred, raises, fated
    "[^aeioulr][lr]e[sd]?$|[csgz]es$|[td]ed$|"
    # fixes miscellaneous issues:
    # flying, piano, video, prism, fire, evaluate
    + ".y[aeiou]|ia(?!n$)|eo|ism$|[^aeiou]ire$|[^gq]ua",
    flags=re.I
)

def count_syllables(word):
    vowel_runs = len(VOWEL_RUNS.findall(word))
    exceptions = len(EXCEPTIONS.findall(word))
    additional = len(ADDITIONAL.findall(word))
    return max(1, vowel_runs - exceptions + additional)

print(count_syllables('Assessed')) 
   #print 2
def total_syllables(x):
  count=0
  for i in x:
    count+=count_syllables(i)
  return count
ee=d['preprocess_txt'].apply(total_syllables)
d['Syllable per Word']=ee/x1  ##total syllables / total words
```
